```python
from scipy.stats import pearsonr

def get_feature_importnace(label):
    for column in data.columns[[54, 55, 56, 57]]:
        if column != label:
            correlation,pvalue = pearsonr(data[column],data[label])
            print('Between {feature} and Kp there is a correlation equal {corr} and p value equal {pval}'.format(feature=column, corr=correlation, pval=pvalue))
    print("end")
```
